#DATA AUGMENTATION


```python
train_datagen = ImageDataGenerator(horizontal_flip=True,                              
                                   vertical_flip=False,     
                                   rescale=1./255, 
                                   validation_split=0.15 
                                   )
#here i save 15% of dataset to validate it in future
```

```python
dataset_path = "./drive/MyDrive/realandfake/"
```

```python
train = train_datagen.flow_from_directory(dataset_path,             
                                          class_mode="binary",          
                                          target_size=(96, 96),       
                                          batch_size=32
                                          , 
                                          subset="training")
#specify the training dataset
```

```python
valid = train_datagen.flow_from_directory(dataset_path,
                                          class_mode="binary",
                                          target_size=(96, 96),
                                          batch_size=32
                                        
                                          ,
                                          subset="validation") 
#specify the validation dataset
```

# MODEL PREPARING
and essentials

```python
MNV2 = MobileNetV2(input_shape=(96, 96, 3),    #neural network   
                          include_top=False,  
                          weights='imagenet'      
                          )  
average_layer = GlobalAveragePooling2D()
model = Sequential([     MNV2,     average_layer,  Dense(256, activation=tf.nn.relu),
                      BatchNormalization(),     Dropout(0.2),     Dense(2, activation=tf.nn.softmax) ])
```

```python
model.layers[0].trainable = False
model.compile(optimizer=Adam(learning_rate=0.001),  
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
model.summary()
```

```python
def scheduler(epoch):
       if epoch <= 2:
        return 0.001
       elif epoch > 2 and epoch <= 15:
        return 0.0001 
       else: 
        return 0.00001 

lr_callbacks = tf.keras.callbacks.LearningRateScheduler(scheduler)
```

```python
temp_val = model.fit_generator(train,
                    epochs=15, 
                    callbacks=[lr_callbacks],
                    validation_data=valid)
```

#ANALYSIS

```python
epochs = 15
train_loss = temp_val.history['loss']
val_loss = temp_val.history['val_loss']
train_acc = temp_val.history['accuracy']
val_acc = temp_val.history['val_accuracy']
range_e = range(epochs)

plt.figure(2,figsize=(6,4))
plt.plot(range_e,train_acc) 
plt.plot(range_e,val_acc)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('train - validation')
plt.grid(True)
plt.legend(['train','valid'],loc=4) 

plt.style.available 
plt.style.use(['classic'])


plt.figure(1,figsize=(6,4))
plt.plot(range_e,train_loss)
plt.plot(range_e,val_loss)

plt.xlabel('Epochs')
plt.ylabel('loss')
plt.title('train - validation')

plt.grid(True)
plt.legend(['train','valid'])
plt.style.available 
plt.style.use(['classic']) 
```
